## Pandas, Funciones y Web Scraping

```python
import pandas as pd
import requests as req
from bs4 import BeautifulSoup
```

```python
def making_soup(url):
    resp = req.get(url)
    print(resp)
    c = resp.content
    soup = BeautifulSoup(c, features='lxml')
    return soup
```

```python
def method_one():
    s = making_soup('https://calculadora-uf.cl/')
    return float(s.find_all('span')[1].text.replace('.','').replace(',','.'))
```

```python
def method_two():
    s = making_soup('https://calculadora-uf.cl/')
    span = s.find_all('span', {'class': 'pages__ValorNumber-ojqwdl-1 idIUwa'})
    return float(span[0].text.replace('.','').replace(',','.'))
```

```python
df = pd.read_csv('data/to_uf.csv')
```

```python
df['price_UF'] = df['price'].apply(lambda row: row/method_two())
```
