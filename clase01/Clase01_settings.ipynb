### 1.2 Mac
- Instalar Xcode desde Itunes: https://itunes.apple.com/us/app/xcode/id497799835?mt=12&ls=1 
- Instalar Brew (package manager): 
    - Abrir Terminal: ```command + space bar > terminal``` luego presiona Enter.
    - Ejecutar en terminal ```/usr/bin/ruby -e "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/master/install)" ```

- Instalar Python3 con Brew:
    - Ejecutar en Terminal ```brew install python3```

- Opcional: PATH environment: set up PATH environment variable, if you used HomeBrew to install Python3, then HomeBrew already added PATH. Do not change PATH environment if you can launch python3 from terminal.
    - Agrega lo siguiente a tu archivo ```~/.profile``` 
    - ```export PATH=/usr/local/bin:/usr/local/sbin:$PATH```
    - Usualmente el directorio es ```PATH="/Library/Frameworks/Python.framework/Versions/3.6/bin:${PATH}" ```

### 2.2 Git

#### 2.2.1 Windows
- Descargar Windows [aquí](https://git-scm.com/download/win) y seguir el paso a paso.

#### 2.2.2 Mac
- En Terminal, ejecutar: ```brew install git```

#### 2.2.3 Linux

- Si usamos DEbian/Ubuntu: ```sudo apt install git-all```

**NOTA**: Para poder configurar nuestras credenciales, en todos los OS, debemos ejecutar en Temrinal/Command Prompt:
- ```git config --global user.name "Emma Paris"```
